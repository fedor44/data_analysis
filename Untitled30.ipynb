```python
import pandas as pd

# 분석을 위한 데이터 로드
real_estate_file = '/content/전체데이터.csv'
df = pd.read_csv(real_estate_file, index_col = False, encoding = 'cp949')

df
```

```python
import numpy as np

# 년도별 거래건수 계산을 위한 전처리
df['날짜'] = pd.to_datetime(df['날짜'])
df['year'] = df['날짜'].dt.year
df['count'] = np.ones(shape=(len(df)), dtype=np.int64)

# 한글 출력을 위한 세팅
#!sudo apt-get install -y fonts-nanum
#!sudo fc-cache -fv
#!rm ~/.cache/matplotlib -rf

import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 

# groupby 함수를 이용한 년도별 매매건수 출력
df.groupby('year')['count'].sum().plot(kind = 'bar')
plt.title('년도별 매매건수')
plt.xlabel('년도')
plt.ylabel('매매건수')
plt.show()
```

```python
df['month'] = df['날짜'].dt.month

# 월별 거래건수
trade = df.groupby(['year', 'month'])['count'].sum().reset_index()

# 월별 평균 기준금리
interest_rate = df.groupby(['year', 'month'])['기준금리'].mean().reset_index()

# 월별 데이터를 추리기 위해 년, 월 컬럼 병합
cols = ['year', 'month']
trade['date'] = trade[cols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)

cols = ['year', 'month']
interest_rate['date'] = interest_rate[cols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)

# 전처리 된 데이터를 이용하여 아파트 거래건수와 기준금리의 데이터들을 그래프로 표현
plt.figure(figsize=(25, 5))

plt.title('아파트 거래건수 및 기준금리', fontsize=16)
plt.xlabel('date')

plt.plot(trade['date'], trade['count'],
         color='skyblue',
         marker='o', markerfacecolor='blue',
         markersize=6)
plt.ylabel('거래건수', color='blue')
plt.tick_params(axis='y', labelcolor='blue')

y_right = plt.twinx()
y_right.plot(interest_rate['date'], interest_rate['기준금리'],
             color='pink',
             marker='o', markerfacecolor='red',
             markersize=6)
plt.ylabel('기준금리', color='red')
y_right.tick_params(axis='y', labelcolor='red')

plt.show()
```

```python
# 평수별 데이터 분석을 위한 전처리
size59 = df[(df['전용면적']>=59) & (df['전용면적']<=60)].copy()

# 월별 아파트 금액 평균
size59['거래금액'] = size59['거래금액'].str.replace(',','')
size59['거래금액'] = size59['거래금액'].astype(int)
size59 = size59.groupby(['year', 'month'])['거래금액'].mean().reset_index()

# 월별 데이터를 추리기 위해 년, 월 컬럼 병합
cols = ['year', 'month']
size59['date'] = size59[cols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)

# 전처리 된 데이터를 이용하여 거래금액과 기준금리의 데이터들을 그래프로 표현
plt.figure(figsize=(25, 5))

plt.title('아파트 거래금액 및 기준금리 (59㎡ 기준)', fontsize=16)
plt.xlabel('date')

plt.plot(size59['date'], size59['거래금액'],
         color='limegreen',
         marker='o', markerfacecolor='green',
         markersize=6)
plt.ylabel('거래금액', color='green')
plt.tick_params(axis='y', labelcolor='green')

y_right = plt.twinx()
y_right.plot(interest_rate['date'], interest_rate['기준금리'],
             color='pink',
             marker='o', markerfacecolor='red',
             markersize=6)
plt.ylabel('기준금리', color='red')
y_right.tick_params(axis='y', labelcolor='red')

plt.show()
```

```python
# 평수별 데이터 분석을 위한 전처리
size84 = df[(df['전용면적']>=84) & (df['전용면적']<=85)].copy()

# 월별 아파트 금액 평균
size84['거래금액'] = size84['거래금액'].str.replace(',','')
size84['거래금액'] = size84['거래금액'].astype(int)
size84 = size84.groupby(['year', 'month'])['거래금액'].mean().reset_index()

# 월별 데이터를 추리기 위해 년, 월 컬럼 병합
cols = ['year', 'month']
size84['date'] = size84[cols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)

# 전처리 된 데이터를 이용하여 거래금액과 기준금리의 데이터들을 그래프로 표현
plt.figure(figsize=(25, 5))

plt.title('아파트 거래금액 및 기준금리 (84㎡ 기준)', fontsize=16)
plt.xlabel('date')

plt.plot(size84['date'], size84['거래금액'],
         color='plum',
         marker='o', markerfacecolor='purple',
         markersize=6)
plt.ylabel('거래금액', color='purple')
plt.tick_params(axis='y', labelcolor='purple')

y_right = plt.twinx()
y_right.plot(interest_rate['date'], interest_rate['기준금리'],
             color='pink',
             marker='o', markerfacecolor='red',
             markersize=6)
plt.ylabel('기준금리', color='red')
y_right.tick_params(axis='y', labelcolor='red')

plt.show()
```

```python
# 월별 주가지수 평균
kospi = df.groupby(['year', 'month'])['KOSPI 지수'].mean().reset_index().copy()

# 월별 데이터를 추리기 위해 년, 월 컬럼 병합
cols = ['year', 'month']
kospi['date'] = kospi[cols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)

# 전처리 된 데이터를 이용하여 거래금액과 KOSPI 지수의 데이터들을 그래프로 표현
plt.figure(figsize=(25, 5))

plt.title('아파트 거래금액 및 KOSPI 지수 (59㎡ 기준)', fontsize=16)
plt.xlabel('date')

plt.plot(size59['date'], size59['거래금액'],
         color='limegreen',
         marker='o', markerfacecolor='green',
         markersize=6)
plt.ylabel('거래금액', color='green')
plt.tick_params(axis='y', labelcolor='green')

y_right = plt.twinx()
y_right.plot(kospi['date'], kospi['KOSPI 지수'],
             color='coral',
             marker='o', markerfacecolor='crimson',
             markersize=6)
plt.ylabel('KOSPI 지수', color='crimson')
y_right.tick_params(axis='y', labelcolor='crimson')

plt.show()
```

```python
plt.figure(figsize=(10, 10))

# 월별 매매수급 평균
trade_index = df.groupby(['year', 'month'])['매매수급동향'].mean().reset_index().copy()

# 월별로 groupby 된 각 데이터 병합
hitmap_df_first = pd.merge(trade, interest_rate)
hitmap_df_second = pd.merge(kospi, trade_index)
hitmap_df = pd.merge(hitmap_df_first, hitmap_df_second)

# 불필요 데이터 제거 및 컬럼명 변경
hitmap_df59 = pd.merge(hitmap_df, size59)

# 각 데이터를 거래금액 기준 산점도로 표현
ax1 = plt.subplot(2, 2, 1)
x1 = hitmap_df59['거래금액']
y1 = hitmap_df59['매매수급동향']
plt.title('거래금액과 매매수급동향(59㎡ 기준)')
plt.scatter(x1, y1)
plt.xlabel('거래금액')
plt.ylabel('매매수급동향')

ax2 = plt.subplot(2, 2, 2)
x2 = hitmap_df59['거래금액']
y2 = hitmap_df59['기준금리']
plt.title('거래금액과 기준금리(59㎡ 기준)')
plt.scatter(x2, y2)
plt.xlabel('거래금액')
plt.ylabel('기준금리')

ax3 = plt.subplot(2, 2, 3)
x3 = hitmap_df59['거래금액']
y3 = hitmap_df59['KOSPI 지수']
plt.title('거래금액과 KOSPI 지수(59㎡ 기준)')
plt.scatter(x3, y3)
plt.xlabel('거래금액')
plt.ylabel('KOSPI 지수')

ax4 = plt.subplot(2, 2, 4)
x4 = hitmap_df59['거래금액']
y4 = hitmap_df59['count']
plt.title('거래금액과 거래건수(59㎡ 기준)')
plt.scatter(x4, y4)
plt.xlabel('거래금액')
plt.ylabel('거래건수')

plt.tight_layout()
plt.show()
```
